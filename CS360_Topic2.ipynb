```python
#Linked List
# Node class
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

# LinkedList class
class LinkedList:
    def __init__(self):
        self.head = None

    # Insert a new node at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    # Insert a new node at the beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Insert after a specific node
    def insert_after_node(self, prev_node_data, data):
        current_node = self.head
        while current_node and current_node.data != prev_node_data:
            current_node = current_node.next
        if not current_node:
            print("The mentioned node is not found.")
            return
        new_node = Node(data)
        new_node.next = current_node.next
        current_node.next = new_node

    # Delete a node by value
    def delete_node(self, key):
        current_node = self.head
        if current_node and current_node.data == key:
            self.head = current_node.next
            current_node = None
            return

        prev_node = None
        while current_node and current_node.data != key:
            prev_node = current_node
            current_node = current_node.next

        if current_node is None:
            print("The node with data", key, "is not found.")
            return

        prev_node.next = current_node.next
        current_node = None

    # Search for a node by value
    def search(self, key):
        current_node = self.head
        while current_node and current_node.data != key:
            current_node = current_node.next
        if current_node:
            return True
        return False

    # Print the linked list
    def print_list(self):
        current_node = self.head
        if current_node is None:
            print("The linked list is empty.")
            return
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next
        print("None")

    # Get the length of the linked list
    def length(self):
        current_node = self.head
        count = 0
        while current_node:
            count += 1
            current_node = current_node.next
        return count

# Example Usage:
if __name__ == "__main__":
    linked_list = LinkedList()

    # Inserting elements
    linked_list.insert_at_end(1)
    linked_list.insert_at_end(2)
    linked_list.insert_at_end(3)
    linked_list.insert_at_beginning(0)

    print("Linked list after insertion:")
    linked_list.print_list()

    # Deleting a node
    linked_list.delete_node(2)
    print("Linked list after deleting node with data 2:")
    linked_list.print_list()

    # Searching for a node
    found = linked_list.search(3)
    print("Node with data 3 found:", found)

    # Length of the linked list
    print("Length of the linked list:", linked_list.length())
```

```python
#Enhanced one

# Node class
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

# LinkedList class with a tail pointer
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None  # Add a tail pointer

    # Insert a new node at the end (O(1) time complexity)
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            self.tail = new_node  # Update tail
        else:
            self.tail.next = new_node  # Link the new node at the end
            self.tail = new_node  # Update tail to the new node

    # Insert a new node at the beginning (O(1) time complexity)
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        if self.tail is None:  # If the list was empty, update the tail
            self.tail = new_node
```

```python
#Array Stacks
class ArrayStack:
    def __init__(self, capacity=10):
        """Initialize the stack with a fixed capacity."""
        self.capacity = capacity
        self.stack = [None] * capacity  # Pre-allocate space for the stack
        self.top = -1  # Index of the top element (-1 means the stack is empty)

    def is_empty(self):
        """Check if the stack is empty."""
        return self.top == -1

    def is_full(self):
        """Check if the stack is full."""
        return self.top == self.capacity - 1

    def push(self, item):
        """Add an item to the stack."""
        if self.is_full():
            raise OverflowError("Stack is full")
        self.top += 1
        self.stack[self.top] = item

    def pop(self):
        """Remove and return the top item from the stack."""
        if self.is_empty():
            raise IndexError("Pop from an empty stack")
        item = self.stack[self.top]
        self.stack[self.top] = None  # Clear the reference
        self.top -= 1
        return item

    def peek(self):
        """Return the top item without removing it."""
        if self.is_empty():
            raise IndexError("Peek from an empty stack")
        return self.stack[self.top]

    def size(self):
        """Return the current size of the stack."""
        return self.top + 1

    def __str__(self):
        """Return a string representation of the stack."""
        return str([self.stack[i] for i in range(self.top + 1)])

# Example usage
if __name__ == "__main__":
    stack = ArrayStack(5)

    # Push elements
    stack.push(10)
    stack.push(20)
    stack.push(30)

    # Check stack state
    print("Stack after pushes:", stack)

    # Peek top element
    print("Peek:", stack.peek())

    # Pop elements
    print("Popped:", stack.pop())

    # Stack state after popping
    print("Stack after pop:", stack)
```

```python
#Linked List Stacks
class Node:
    """Node class to store data and the next node reference."""
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedListStack:
    def __init__(self):
        """Initialize an empty stack."""
        self.top = None  # Top points to the head node of the linked list
        self._size = 0   # Keep track of the stack size

    def is_empty(self):
        """Check if the stack is empty."""
        return self.top is None

    def push(self, item):
        """Add an item to the top of the stack."""
        new_node = Node(item)
        new_node.next = self.top  # The new node points to the current top
        self.top = new_node  # Update top to the new node
        self._size += 1

    def pop(self):
        """Remove and return the top item from the stack."""
        if self.is_empty():
            raise IndexError("Pop from an empty stack")
        item = self.top.data
        self.top = self.top.next  # Move the top to the next node
        self._size -= 1
        return item

    def peek(self):
        """Return the top item without removing it."""
        if self.is_empty():
            raise IndexError("Peek from an empty stack")
        return self.top.data

    def size(self):
        """Return the current size of the stack."""
        return self._size

    def __str__(self):
        """Return a string representation of the stack."""
        result = []
        current = self.top
        while current:
            result.append(current.data)
            current = current.next
        return str(result)

# Example usage
if __name__ == "__main__":
    stack = LinkedListStack()

    # Push elements
    stack.push(10)
    stack.push(20)
    stack.push(30)

    # Check stack state
    print("Stack after pushes:", stack)

    # Peek top element
    print("Peek:", stack.peek())

    # Pop elements
    print("Popped:", stack.pop())

    # Stack state after popping
    print("Stack after pop:", stack)
```

```python
# Linked List Queue
class Node:
    def __init__(self, data):
        # Each node has data and a reference to the next node
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        # Initialize the front and rear pointers of the queue
        self.front = None
        self.rear = None

    def is_empty(self):
        # Return True if the queue is empty, False otherwise
        return self.front is None

    def enqueue(self, item):
        # Create a new node for the new item
        new_node = Node(item)
        if self.rear is None:
            # If the queue is empty, front and rear are the same
            self.front = self.rear = new_node
        else:
            # Add the new node at the end of the queue
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        # Remove and return the item from the front of the queue
        if self.is_empty():
            raise IndexError("Dequeue from an empty queue")

        temp = self.front
        self.front = temp.next

        if self.front is None:
            # If the queue is now empty, set rear to None as well
            self.rear = None

        return temp.data

    def peek(self):
        # Return the front element without removing it
        if self.is_empty():
            raise IndexError("Peek from an empty queue")
        return self.front.data

    def display(self):
        # Display the queue elements
        if self.is_empty():
            print("Queue is empty")
            return

        current = self.front
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Example usage:
queue = Queue()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
queue.display()

print("Dequeue:", queue.dequeue())
queue.display()
```

```python
#Doubly Linked List
class Node:
    def __init__(self, data):
        # Each node contains data, a reference to the next node, and a reference to the previous node
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        # Initialize the doubly linked list with head and tail pointers
        self.head = None
        self.tail = None

    def is_empty(self):
        # Check if the list is empty
        return self.head is None

    def append(self, data):
        # Add a new node at the end (tail) of the list
        new_node = Node(data)
        if self.is_empty():
            # If the list is empty, the new node becomes both head and tail
            self.head = self.tail = new_node
        else:
            # Update the links to add the new node at the tail
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node  # Move the tail to the new node

    def prepend(self, data):
        # Add a new node at the beginning (head) of the list
        new_node = Node(data)
        if self.is_empty():
            # If the list is empty, the new node becomes both head and tail
            self.head = self.tail = new_node
        else:
            # Update links to add the new node before the current head
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node  # Move the head to the new node

    def delete(self, data):
        # Delete a node with the given data
        if self.is_empty():
            print("List is empty, nothing to delete.")
            return

        current = self.head
        while current:
            if current.data == data:
                # If the node to be deleted is the head
                if current == self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                    else:
                        # If the list becomes empty, reset the tail
                        self.tail = None
                # If the node to be deleted is the tail
                elif current == self.tail:
                    self.tail = current.prev
                    self.tail.next = None
                else:
                    # Update the links of surrounding nodes to skip the current node
                    current.prev.next = current.next
                    current.next.prev = current.prev
                return
            current = current.next
        print(f"Node with data {data} not found.")

    def display(self):
        # Display the list from head to tail
        if self.is_empty():
            print("List is empty.")
            return

        current = self.head
        while current:
            print(current.data, end=" <-> " if current.next else "\n")
            current = current.next

    def reverse_display(self):
        # Display the list from tail to head
        if self.is_empty():
            print("List is empty.")
            return

        current = self.tail
        while current:
            print(current.data, end=" <-> " if current.prev else "\n")
            current = current.prev

# Example usage:
dll = DoublyLinkedList()
dll.append(10)
dll.append(20)
dll.prepend(5)
dll.display()

dll.delete(20)
dll.display()

dll.reverse_display()
```

```python
class CircularSequence:
    def __init__(self, capacity):
        self.capacity = capacity  # Maximum size of the circular array
        self.array = [None] * capacity
        self.n = 0  # Number of elements in the sequence
        self.f = 0  # Index of the first element
        self.l = 0  # Index of the position after the last element

    def rank2index(self, rank):
        """Convert rank to index using the modulo operator."""
        return (self.f + rank) % self.capacity

    def index2rank(self, index):
        """Convert index to rank using the modulo operator."""
        return (self.capacity - self.f + index) % self.capacity

    def elemAtRank(self, rank):
        """Return the element at the given rank without removing it."""
        if rank < 0 or rank >= self.n:
            raise IndexError("Rank out of bounds")
        index = self.rank2index(rank)
        return self.array[index]

    def size(self):
        """Return the current size of the sequence."""
        return self.n

    def isEmpty(self):
        """Check if the sequence is empty."""
        return self.n == 0

    def insertLast(self, element):
        """Insert an element at the last position."""
        if self.size() == self.capacity - 1:
            raise Exception("Sequence is full")
        self.array[self.l] = element
        self.l = (self.l + 1) % self.capacity
        self.n += 1

    def insertFirst(self, element):
        """Insert an element at the first position."""
        if self.size() == self.capacity - 1:
            raise Exception("Sequence is full")
        self.f = (self.capacity + self.f - 1) % self.capacity
        self.array[self.f] = element
        self.n += 1

    def insertAtRank(self, rank, element):
        """Insert an element at the given rank, shifting elements efficiently."""
        if rank < 0 or rank > self.n:
            raise IndexError("Rank out of bounds")
        if self.size() == self.capacity - 1:
            raise Exception("Sequence is full")

        if rank == 0:
            self.insertFirst(element)
        elif rank == self.n:
            self.insertLast(element)
        else:
            index = self.rank2index(rank)
            if rank < self.n // 2:
                # Shift elements left towards the front
                self.f = (self.f - 1 + self.capacity) % self.capacity
                for i in range(0, rank):
                    self.array[self.rank2index(i)] = self.array[self.rank2index(i + 1)]
            else:
                # Shift elements right towards the end
                for i in range(self.n, rank, -1):
                    self.array[self.rank2index(i)] = self.array[self.rank2index(i - 1)]
                self.l = (self.l + 1) % self.capacity
            self.array[index] = element
            self.n += 1

    def removeAtRank(self, rank):
        """Remove the element at the given rank, shifting elements efficiently."""
        if rank < 0 or rank >= self.n:
            raise IndexError("Rank out of bounds")

        index = self.rank2index(rank)
        removed_element = self.array[index]

        if rank < self.n // 2:
            # Shift elements right towards the end (towards front)
            for i in range(rank, 0, -1):
                self.array[self.rank2index(i)] = self.array[self.rank2index(i - 1)]
                #print('done')
            self.f = (self.f + 1) % self.capacity
        else:
            # Shift elements left towards the front
            for i in range(rank, self.n - 1):
                self.array[self.rank2index(i)] = self.array[self.rank2index(i + 1)]
            self.l = (self.l - 1 + self.capacity) % self.capacity

        self.n -= 1
        return removed_element

# Testing the CircularSequence class with efficient insertion and removal

sequence = CircularSequence(10)

# Insert elements at the first and last positions
sequence.insertFirst(10)
sequence.insertLast(20)
sequence.insertLast(30)

# Insert element at rank 1 (between 10 and 20)
sequence.insertAtRank(1, 15)

# Test the contents of the sequence
print("Sequence after insertion:", [sequence.elemAtRank(i) for i in range(sequence.size())])

# Remove the element at rank 1
removed = sequence.removeAtRank(1)
print(f"Removed element: {removed}")

# Test the contents of the sequence after removal
print("Sequence after removal:", [sequence.elemAtRank(i) for i in range(sequence.size())])
```
