```python
def quickSort(arr):
    # Base case: If the array has 0 or 1 elements, it's already sorted
    if len(arr) <= 1:
        return arr

    # Step 1: Choose a pivot (here, we choose the middle element)
    pivot = arr[len(arr) // 2]

    # Step 2: Partition into three lists
    smaller = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    larger = [x for x in arr if x > pivot]

    # Step 3: Recursively sort the smaller and larger lists, then concatenate
    return quickSort(smaller) + equal + quickSort(larger)

# Example usage:
arr = [10, 7, 8, 9, 1, 5]
sorted_arr = quickSort(arr)
print("Sorted array:", sorted_arr)
```

```python
import random

def in_place_quick_sort(S, l, r):

    if l < r:
        p = in_place_partition(S, l, r)
        in_place_quick_sort(S, l, p - 1)
        in_place_quick_sort(S, p + 1, r)


def in_place_partition(S, lo, hi):


    # p ← a random integer between lo and hi
    p = random.randint(lo, hi)

    # swapElements(S, lo, p)
    S[lo], S[p] = S[p], S[lo]

    # pivot ← S.elemAtRank(lo)
    pivot = S[lo]

    # j ← lo + 1, k ← hi
    j = lo + 1
    k = hi

    # while j < k do
    while j < k:
        # while j < k ∧ S[j] < pivot do j ← j + 1
        while j < k and S[j] < pivot:
            j += 1

        # while k > j ∧ S[k] > pivot do k ← k - 1
        while k > j and S[k] > pivot:
            k -= 1

        # if j < k then swapElements(S, j, k); j++, k--
        if j < k:
            S[j], S[k] = S[k], S[j]
            j += 1
            k -= 1

    # swapElements(S, lo, k)  // move pivot to sorted rank
    S[lo], S[k] = S[k], S[lo]

    # return k
    return k


# ---- Example usage ----
if __name__ == "__main__":
    arr = [10, 7, 8, 9, 1, 5]
    print("Original array:", arr)
    in_place_quick_sort(arr, 0, len(arr) - 1)
    print("Sorted array:", arr)
```
