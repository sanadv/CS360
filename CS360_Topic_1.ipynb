```python
#ArrayOperations
class ArrayOperations:
    def __init__(self, array):
        self.array = array

    def find_max(self):
        max_value = self.array[0]
        for num in self.array:
            if num > max_value:
                max_value = num
        return max_value

    def find_min(self):
        min_value = self.array[0]
        for num in self.array:
            if num < min_value:
                min_value = num
        return min_value

    def find_average(self):
        total = 0
        for num in self.array:
            total += num
        return total / len(self.array)

    def find_sum(self):
        total = 0
        for num in self.array:
            total += num
        return total

    def find_product(self):
        product = 1
        for num in self.array:
            product *= num
        return product

    def count_elements(self):
        count = 0
        for _ in self.array:
            count += 1
        return count

    def reverse_array(self):
        reversed_array = []
        for i in range(len(self.array)-1, -1, -1):
            reversed_array.append(self.array[i])
        return reversed_array

    def is_sorted(self):
        for i in range(len(self.array) - 1):
            if self.array[i] > self.array[i + 1]:
                return False
        return True

    def find_element(self, element):
        for i in range(len(self.array)):
            if self.array[i] == element:
                return i
        return -1

    def find_median(self):
        sorted_array = sorted(self.array)
        n = len(sorted_array)
        middle = n // 2
        if n % 2 == 0:
            median = (sorted_array[middle - 1] + sorted_array[middle]) / 2
        else:
            median = sorted_array[middle]
        return median

    def remove_duplicates(self):
        unique_elements = []
        for num in self.array:
            if num not in unique_elements:
                unique_elements.append(num)
        return unique_elements

    def find_second_largest(self):
        first = second = float('-inf')
        for num in self.array:
            if num > first:
                second = first
                first = num
            elif first > num > second:
                second = num
        return second

    def rotate_left(self, k):
        n = len(self.array)
        rotated_array = []
        for i in range(n):
            rotated_array.append(self.array[(i + k) % n])
        return rotated_array

# Example usage:
array = [3, 1, 4, 1, 5, 9, 2, 6, 5]
operations = ArrayOperations(array)
# Function calls and prints

max_value = operations.find_max()
print(f"Max: {max_value}")

min_value = operations.find_min()
print(f"Min: {min_value}")

average = operations.find_average()
print(f"Average: {average}")

sum_value = operations.find_sum()
print(f"Sum: {sum_value}")

product_value = operations.find_product()
print(f"Product: {product_value}")

count = operations.count_elements()
print(f"Count of Elements: {count}")

reversed_array = operations.reverse_array()
print(f"Reversed Array: {reversed_array}")

is_sorted = operations.is_sorted()
print(f"Is Sorted: {is_sorted}")

element_index = operations.find_element(5)
print(f"Find Element (5): {element_index}")

median = operations.find_median()
print(f"Median: {median}")

unique_array = operations.remove_duplicates()
print(f"Remove Duplicates: {unique_array}")

second_largest = operations.find_second_largest()
print(f"Second Largest: {second_largest}")

rotated_array = operations.rotate_left(2)
print(f"Rotate Left (by 2): {rotated_array}")
```

```python
#Sum of First n Natural Numbers
class SumOfNumbers:
    def __init__(self, n):
        self.n = n

    def sum_using_loop(self):
        total = 0
        for i in range(1, self.n + 1):
            total += i
        return total

    def sum_using_formula(self):
        return self.n * (self.n + 1) // 2

    def measure_execution_time(self, method):
        start_time = time.time()
        result = method()
        end_time = time.time()
        execution_time = end_time - start_time
        return result, execution_time

    def format_time(self, time_seconds):
        if time_seconds < 1e-3:
            return f"{time_seconds * 1e6:.2f} µs"
        elif time_seconds < 1:
            return f"{time_seconds * 1e3:.2f} ms"
        else:
            return f"{time_seconds:.2f} s"

    def compare_algorithms(self):
        result_loop, time_loop = self.measure_execution_time(self.sum_using_loop)
        result_formula, time_formula = self.measure_execution_time(self.sum_using_formula)

        print(f"Sum using loop: {result_loop}, Time taken: {self.format_time(time_loop)}")
        print(f"Sum using formula: {result_formula}, Time taken: {self.format_time(time_formula)}")

# Example usage:
n = 10**7
sum_of_numbers = SumOfNumbers(n)
sum_of_numbers.compare_algorithms()

```
